```python
#Exercise 1: Prime Number
def is_prime(number):
    # A prime number is greater than 1
    if number <= 1:
        return False
    # 2 and 3 are prime numbers
    if number <= 3:
        return True
    # Eliminate even numbers and numbers divisible by 3
    if number % 2 == 0 or number % 3 == 0:
        return False
    # Check for factors from 5 up to the square root of the number
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

# Test the function
number = int(input("Enter a number: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
#Exercise 2: Product of Random Numbers
import random

def multiply_quiz():
    # Generate two random numbers between 1 and 10
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)

    # Calculate the correct product
    correct_answer = num1 * num2

    # Ask the user to enter the product
    user_answer = int(input(f"What is {num1} * {num2}? "))

    # Check if the user's answer is correct
    if user_answer == correct_answer:
        print("Correct! Well done!")
    else:
        print(f"Incorrect. The correct answer is {correct_answer}.")

# Run the multiplication quiz
multiply_quiz()
```

```python
#Exercise 3: Squares of Even/Odd Numbers.
def print_odd_squares():
    for number in range(100, 201):
        if number % 2 != 0:
            print(f"The square of {number} is {number ** 2}")

def print_even_squares():
    for number in range(100, 201):
        if number % 2 == 0:
            print(f"The square of {number} is {number ** 2}")

def switch_case(choice):
    switcher = {
        "odd": print_odd_squares,
        "even": print_even_squares
    }
    # Get the function from the dictionary and call it
    func = switcher.get(choice, lambda: print("Invalid choice"))
    func()

choice = "odd"

switch_case(choice)
```

```python
#Exercise 4: Word counter
from collections import Counter

def count_words(input_text):
    # Split the text into words based on whitespace
    words = input_text.split()

    # Use Counter to count occurrences of each word
    word_count = Counter(words)

    # Print the word counts
    for word, count in word_count.items():
        print(f"'{word}': {count}")

# Example usage
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
count_words(input_text)
```

```python
#Exercise 5: Check for Palindrome
import string

def is_palindrome(s):
    # Remove spaces and punctuation, and convert to lowercase
    normalized_string = ''.join(c.lower() for c in s if c.isalnum())

    # Check if the normalized string reads the same forward and backward
    return normalized_string == normalized_string[::-1]

# Example usage
input_string = "racecar"
print(is_palindrome(input_string))
```
