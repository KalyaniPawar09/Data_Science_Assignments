```python
#3
#---HISTPLOT
import matplotlib.pyplot as plt

# List of numerical columns to plot
numerical_columns = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV']

# Create histograms for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    df[column].hist(bins=30)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#----BOXPLOT
import seaborn as sns

# Create boxplots for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
#----PIECHART
plt.figure(figsize=(10, 6))
df['LB'].value_counts().plot(kind='pie', autopct='%1.1f%%') # LB should be a string
plt.title('Distribution of Tendency')
plt.ylabel('')
plt.show()
```

```python
#---SCATTER PLOTS
# Example scatter plot between two numerical columns
plt.figure(figsize=(10, 6))
plt.scatter(df['LB'], df['MSTV'])
plt.title('Scatter Plot of LB vs MSTV')
plt.xlabel('LB (Baseline Fetal Heart Rate)')
plt.ylabel('MSTV (Mean Value of Short Term Variability)')
plt.show()
```

```python
#----# correlation matrix
correlation_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#---- pair plot for selected numerical columns
sns.pairplot(df[['LB', 'AC', 'FM', 'UC', 'DL', 'DS']])
plt.title('Pair Plot of Selected Variables')
plt.show()
```

```python
#4
print(df.head())

# Example of plotting trends if a 'Date' column exists
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)

    # Plot trends for numerical columns
    for column in df.columns:
        if df[column].dtype in ['float64', 'int64']:  # Only plot numerical columns
            plt.figure(figsize=(12, 6))
            df[column].plot()
            plt.title(f'Trend of {column} Over Time')
            plt.xlabel('Date')
            plt.ylabel(column)
            plt.show()
else:
    print("\n No DATE column found in the dataset.")
```

```python
# - The correlation matrix shows strong positive correlations between 'ASTV' and 'MSTV',
#   which may indicate that as the percentage of abnormal short-term variability increases,
#   the mean value of short-term variability also increases.
# - A strong negative correlation between 'DL' (Late Decelerations) and 'AC' (Accelerations)
#   suggests that higher late decelerations are associated with fewer accelerations, potentially indicating fetal distress.
# - Trends over time reveal increasing trends in 'MSTV', suggesting that mean short-term variability is increasing over the observed period.
# - No temporal data is available
```
