```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
#Loading The Dataset
data = pd.read_csv('/content/wine.csv')
```

```python
# Task 1: Exploratory Data Analysis (EDA)
# 1. Basic Data Exploration
print(data.info())
```

```python
print(data.describe())
```

```python
# 2. Visualize feature distributions
#Histplots
data.hist(bins=15, figsize=(15, 10), color='blue', edgecolor='black')
plt.suptitle('Feature Distributions', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Box plots
for column in data.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=data[column], color='cyan')
    plt.title(f'Box Plot of {column}', fontsize=14)
    plt.show()
```

```python
#Density Plots
for column in data.columns:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data[column], shade=True, color='green')
    plt.title(f'Density Plot of {column}', fontsize=14)
    plt.show()
```

```python
#Correlation Matrix
plt.figure(figsize=(12,8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Features', fontsize=16)
plt.show()
```

```python
# Task 2: Dimensionality Reduction with PCA
# 1. Standardize the features
features = data.drop('Type', axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# 2. Apply PCA
pca = PCA()
pca.fit(scaled_features)
```

```python
# 3. Scree Plot
plt.figure(figsize=(10,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Components', fontsize=14)
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
# 4. Apply PCA with selected number of components (e.g., 5)
pca = PCA(n_components=5)
pca_transformed_data = pca.fit_transform(scaled_features)
```

```python
# Task 3: Clustering on Original Data
# 1. Apply KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_features)
```

```python
# 2. Evaluate clustering performance
silhouette_orig = silhouette_score(scaled_features, kmeans_labels)
db_score_orig = davies_bouldin_score(scaled_features, kmeans_labels)
```

```python
print(f'Silhouette Score (Original Data): {silhouette_orig}')
print(f'Davies-Bouldin Score (Original Data): {db_score_orig}')
```

```python
# 3. Visualize clusters
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=kmeans_labels, cmap='viridis', edgecolor='k')
plt.title('KMeans Clustering on Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Task 4: Clustering on PCA-Transformed Data
# 1. Apply KMeans to PCA-Transformed Data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_labels_pca = kmeans_pca.fit_predict(pca_transformed_data)
```

```python
# 2. Evaluate clustering performance on PCA data
silhouette_pca = silhouette_score(pca_transformed_data, kmeans_labels_pca)
db_score_pca = davies_bouldin_score(pca_transformed_data, kmeans_labels_pca)

print(f'Silhouette Score (PCA Data): {silhouette_pca}')
print(f'Davies-Bouldin Score (PCA Data): {db_score_pca}')
```

```python
# 3. Visualize PCA Clustering
plt.scatter(pca_transformed_data[:, 0], pca_transformed_data[:, 1], c=kmeans_labels_pca, cmap='viridis', edgecolor='k')
plt.title('KMeans Clustering on PCA-Transformed Data')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
# Task 5: Comparison and Analysis
print(f'Comparison of Clustering Performance:\n'
      f'Original Data - Silhouette: {silhouette_orig}, Davies-Bouldin: {db_score_orig}\n'
      f'PCA Data - Silhouette: {silhouette_pca}, Davies-Bouldin: {db_score_pca}')
```
