Chi square test

```python
#2
import scipy.stats as stats
import numpy as np
satisfaction = np.array([[50, 70],  # Very satisfied
                     [80, 100],  # Satisfied
                     [60, 90],   # Neutral
                     [30, 50],   # Unsatisfied
                     [20, 50]])  # Very unsatisfied
```

```python
chi2_stat,p_val,dof,expected=stats.chi2_contingency(satisfaction)
print("Chi-Square satistic (X2):",chi2_stat)
print("P-value:",p_val)
print("Degrees of freedom:",dof)
print("Expected frequencies:\n",expected)
```

```python
#3
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, dof)
print("Critical Value:", critical_value)
```

```python
#4
if chi2_stat > critical_value:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

# Display the decision
print("Decision:", decision)
```

```python
#1 as we fail to reject the null hypothesis,The state of hypothesis is null hypothesis
```

```python
from scipy.stats import t
import math
```

```python
samplemean = 3050  # Sample mean
theoreticalmean = 4000  # Theoretical mean
standarddev = 5 * 25  # given Standard deviation of the
samplesize = 25  # given Sample size
alpha = 0.05  #given Significance level
```

```python
#2
test_statistic = (samplemean - theoreticalmean) / (standarddev / math.sqrt(samplesize))
print("test statistic value :",test_statistic)
```

```python
#3
dof=samplesize-1
print("degree of freedom :",dof)
critical_value=t.ppf(1-alpha,dof)
print("critical value :",critical_value)
```

```python
#4
if test_statistic < critical_value:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

print("Decision:", decision)
```
